```python
import requests
import json
import os
import pandas as pd
from bs4 import BeautifulSoup


def strip_html(html_string):
    soup = BeautifulSoup(html_string, "html.parser")
    return soup.get_text()

baseurl = "https://hvl.pureservice.com/agent/api"

PStoken = os.environ["PStoken"]
headers = {'X-Authorization-Key': PStoken, 'Content-Type': 'application/vnd.api+json'}
```

```python
dataliste = []
url = f'{baseurl}/user/?filter=!Disabled %26%26 (role == "Agent")&sort=firstName ASC,lastName ASC'
response = requests.request("GET", url, headers=headers)
print(response.status_code)
if response.status_code == 200:
    print(response.json())
    for user in response.json()['users']:
        dataliste.append(pd.DataFrame([user]))
alledata = pd.concat(dataliste)
```

```python
alledata.to_csv("brukarar.csv", index=False)
```

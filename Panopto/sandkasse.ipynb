```python
import pandas as pd

def splitt(brukarnamn):
    streng = str(brukarnamn)
    if '@' in streng:
        domene = streng.split('@')[1]
        brukar = streng.split('\\')[1].split('@')[0]
        if domene == "hvl.no":
            if brukar.isdigit():
                return "Student"
            else:
                return "Tilsett"
        else:
            return "Ekstern"
    else:
        return " "#
```

```python
data = pd.read_csv("totalsessionscreatedperuser.csv")
data['Rolle'] = data['UserName'].apply(splitt)
```

```python
data['Gjennomsnitt'] = (data['Minutes Created'].str.replace(',', '.').astype(float)/data['Number of Sessions Created'].astype(float)).round(1)
```

```python
data.head(10)
```

```python
data['Brukar'] = data['UserName'].str.split('\\').str[1]
```

```python
data['HVL'] = data['UserName'].str.split('@').str[1]=="hvl.no"
```

```python
data['Minutt'] = data['Minutes Created'].str.replace(',', '.').astype(float)
```

```python
data['Antall'] = data['Number of Sessions Created'].astype(int)
```

```python
data[['Email', 'Brukar', 'HVL', 'Minutt', 'Antall']].to_csv("statistikk.csv", index=False)
```

```python
import requests
import json
import aasmund
import pandas as pd
from bs4 import BeautifulSoup

baseurl = "https://hvl.pureservice.com/agent/api"

PStoken = aasmund.PStoken
headers = {'X-Authorization-Key': PStoken, 'Content-Type': 'application/vnd.api+json'}

```

```python
url = f"{baseurl}/status?limit=100"
response = requests.request("GET", url, headers=headers)
```

```python
status_liste = []
for s in response.json()['statuses']:
    status_liste.append(pd.DataFrame(s))
status_df = pd.concat(status_liste)
```

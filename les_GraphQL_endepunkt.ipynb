# Les eit GraphQL-endepunkt
Denne gangen les eg alle påmeldingar til alle studentar i eit gitt emne. For å finne kor mange som høyrer til i eit Campusrom.

# 

```python
query = """
query MyQuery($emneid: ID!) {
  course(id: $emneid) {
    enrollmentsConnection {
      nodes {
        user {
          sisId
          enrollments {
            course {
              name
            }
          }
        }
      }
    }
  }
}
"""
```

```python
dat153 = akvut.query_canvas_graphql(query, {'emneid': '28911'})
```

For å kunne tolke svaret frå GraphQl bør eg liste opp dei første verdiane:

```python
dat153
```

Så kan eg plukke ut ei liste over studentar, og kva campusrom dei høyrer til i:

```python
for student in dat153['data']['course']['enrollmentsConnection']['nodes']:
    for emne in student['user']['enrollments']:
        if "Campusrom" in emne['course']['name']:
            print(student['user']['sisId'], emne['course']['name'])
```

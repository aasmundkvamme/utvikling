# Redigere emnemeny

Eg lister først ut alle meny-element i eit eksisterande emne:

```python
emne = 28643
url = f"https://hvl.instructure.com/api/v1/courses/{emne}/tabs"

respons = requests.get(url, headers=headers, params=params)
if 200 <= respons.status_code < 300:
    data = respons.json()
df = pd.DataFrame(data)
df.to_csv("meny.csv", index=False)
```

Her er det tre parametre som styrer plassering/synlighet:

1. "position": Kvar i menyen meny-elementet skal finnast. Dersom det står seinare enn "Innstillingar" ("settings"), så er elementet skjult for studentar.
2. "hidden": Vert satt til TRUE når elementet er skjult (altså plassert seinare enn "Innstillingar" ("settings")).
3. "visibility": Kven meny-elementet skal vere synleg for.

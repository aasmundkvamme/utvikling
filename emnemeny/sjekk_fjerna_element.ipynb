```python
import requests
import os
import pandas as pd
tokenCanvas = os.environ['tokenCanvas']

headers = {
    'Authorization': f'Bearer {tokenCanvas}',
}
params = {
    'per_page': 100,
}
```

```python
data = pd.read_csv("courses.csv")
data.head()
```

```python
emneliste = data['canvas_course_id'].unique().tolist()
fjerna_liste = []
for i in emneliste:
    url = f"https://hvl.instructure.com/api/v1/courses/{i}/tabs"
    respons = requests.get(url, headers=headers, params=params)
    if 200 <= respons.status_code < 300:
        meny = respons.json()
        for m in meny:
            if m['id'] == 'context_external_tool_55':
                if 'hidden' in m:
                    fjerna_liste.append({"canvas_course_id": i, "account_id": data[data['canvas_course_id'] == i]['account_id'].values[0], "hidden": m['hidden']})
```

```python
pd.DataFrame(fjerna_liste).to_csv("fjerna_liste.csv", index=False)
```

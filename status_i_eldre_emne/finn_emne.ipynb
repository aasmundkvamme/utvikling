```python
import pandas as pd
emne = pd.read_csv("emne.csv")
emne = emne[(emne['canvas_term_id'] > 1)&(emne['status'] == 'active')]
emne.head()
```

```python
terminar = pd.read_csv("terminar_rydda.csv", sep=';')
terminar.head()
```

```python
aktuelle_terminar= terminar[terminar['Slutt'] <="2023,5"]['id']
```

```python
aktuelle_emne = emne[emne['canvas_term_id'].isin(aktuelle_terminar)]
```

```python
aktuelle_emne.to_csv('aktuelle_emne.csv', index=False)
```

# Setje eldre påmeldingar til "concluded"

Eg skal gå gjennom ein haug med eldre emne, og finn alle "enrollments" som ikkje er Teststudent, og som er aktive. Desse kan eg så setje til "concluded".

```python
data = pd.read_csv("aktuelle_emne.csv")
sjekk_emne = data['canvas_course_id']
```

```python
deaktiver_desse = []
for course_id in sjekk_emne:
    url = f"https://hvl.instructure.com/api/v1/courses/{course_id}/enrollments"
    print(url)
    respons = requests.get(url, headers=headers, params=params)
    data = respons.json()
    for d in data:
        if (d['enrollment_state']=='active')&(d['user']['name']!="Teststudent"):
            if d['role']=='StudentEnrollment':
                deaktiver_desse.append(f"{d['user']['sis_user_id']},{d['sis_course_id']},student,completed")
            elif d['role']=='TeacherEnrollment':
                deaktiver_desse.append(f"{d['user']['sis_user_id']},{d['sis_course_id']},teacher,completed")
            else:
                deaktiver_desse.append(f"{d['user']['sis_user_id']},{d['sis_course_id']},other,completed")
```

```python
pd.DataFrame(deaktiver_desse).to_csv("deaktiver_desse.csv", index=False)
```

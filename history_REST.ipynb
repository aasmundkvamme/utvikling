```python
import requests
import pandas as pd
import os
```

```python

parametreCanvas = {'per_page': '100'}
hodeCanvas = {'Authorization': 'Bearer ' + os.environ["tokenCanvas"]}

# user_ids = [17586,  37725,  34716,  34723,  31518,  18741,  34512,  33167,  21395,  1574,  8131,  24895,  38862,  44761,  19280,  33849,  49121,  47465,  49148,  49854]
user_ids = [17586]
for user_id in user_ids:
    url = f"https://hvl.instructure.com/api/v1/users/{user_id}/history"
    respons = requests.get(url, headers=hodeCanvas, params=parametreCanvas)
    if 200 <= respons.status_code < 300:
        temp = respons.json()
        if user_id == user_ids[0]:
            data = pd.DataFrame(temp)
        else:
            data = pd.concat([data, pd.DataFrame(temp)])
        print(f"Antal linjer: {len(data)}")
```

```python
data.to_csv("history_REST.csv", index=False)
```
